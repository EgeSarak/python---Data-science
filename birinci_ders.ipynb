```python
import datetime

#: Adding a new column with all values as `0`
#! df['q'] = 0
#! print(df['q'])

# sign(x) 
# x >  0  => 1
# x == 0  => 0
# x <  0  => -1

# pos(x) =
# x >  0 => 1
# x == 0 => 0

def pos(x):
    if x > 0: 
        return 1
    return 0

def pos2(x):
    return 1 if x > 0 else 0

#: Lambda function
def pos3(x):
    if x > 0:
        return 1
    else:
        if x == 0:
            return 0
        else:
            return -1

poslambda = lambda x: 1 if x > 0 else 0

YEAR = datetime.datetime.now().year # 2022
agelambda = lambda x: YEAR - x

   


# APPLY = FOREACH # HER BIRI ICIN
df['renewed'] = df['yr_renovated'].apply( poslambda )

df['age'] = df['yr_built'].apply( agelambda )
df['age2'] = YEAR - df['yr_built']

#: Sorting dataset
#: Siralama
df = df.sort_values(['price'])

#: Filtering


# df[  'columnname'   ] => o kolonu secip degerlerini donduruyor
# df[ ['x', 'y'] ] => o kolonlari secip degerlerini donduruyor



df = df[      df['price'] > 50000       ]






# yr_built
# age
# 2022



# yr_renovated


#! columnlist = list(df.columns)
#! columnlist.remove( 'renewed' )
#! columnlist.insert( 7, 'renewed')
#! df = df[ columnlist ]

#: Saving to a file
df.to_csv("data2.csv", index = None)
```
